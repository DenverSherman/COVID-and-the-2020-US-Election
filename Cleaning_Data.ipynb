```python
# Set up dependencies
import pandas as pd
```

# Cleaning the COVID19 data

```python
# Reading in COVID data
covid_data_raw = r"COVID_DATA\US_counties_COVID19_health_weather_data.csv"
covid_counties_df = pd.read_csv(covid_data_raw)
covid_counties_df.head()
```

```python
# looking at shape of covid data
covid_counties_df.shape
```

```python
covid_counties_df.dtypes
```

```python
# Examining columns currently in DF
covid_counties_df.columns.to_list()
```

```python
# Keeping only certain columns
covid_counties_df = covid_counties_df[[
    "date",
    "county",
    "state",
    "fips",
    "cases",
    "deaths",
    "stay_at_home_announced",
    "stay_at_home_effective",
    "total_population",
    "population_density_per_sqmi",
    "percent_single_parent_households_CHR",
    "percent_adults_with_diabetes",
    "percent_food_insecure",
    "percent_insufficient_sleep",
    "percent_unemployed_CDC",
    "percent_no_highschool_diploma",
    "percent_age_65_and_older",
    "percent_age_17_and_younger",
    "percent_disabled",
    "percent_multi_unit_housing",
    "percent_fair_or_poor_health",
    "percent_not_proficient_in_english",
    "percent_female",
    "percent_rural",
    "percent_minorities"
]]
covid_counties_df
```

```python
covid_counties_df.shape
```

```python
# Dropping columns with NAs
covid_counties_df.dropna(inplace=True)
covid_counties_df.shape
```

```python
# Dropping fips rows that are not numbers
covid_counties_df = covid_counties_df[covid_counties_df["fips"].apply(lambda x: x.isnumeric())]
covid_counties_df.dtypes
```

```python
# Converting fips into integers
covid_counties_df["fips"] = covid_counties_df["fips"].astype(str).astype(int)
covid_counties_df.dtypes
```

```python
# Converting date to datetime data type
covid_counties_df["date"] = pd.to_datetime(covid_counties_df["date"])
covid_counties_df.dtypes
```

```python
# Exporting cleaned COVID data as csv
covid_counties_df.to_csv(r"Clean_data\clean_COVID_data.csv")
```
